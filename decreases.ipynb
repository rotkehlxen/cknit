## Abnahmen gleichmäßig verteilen

Abnahmen werden gemacht, in dem jeweils 2 M zusammengestrickt werden. Wenn wir mit `m` Maschen starten und `a` Maschen abnehmen wollen, dann bleiben also `k = m - 2a` Maschen "unberührt", werden also ganz normal dem Muster entsprechend abgestrickt. Wie verteilen wir jetzt diese Abnahmen am besten? 

Ein Beispiel. Wir haben 57 M und wollen 11 M gleichmäßig verteilt abnehmen:

```python
m = 57
a = 11

k = m - 2*a

print(k)
```

Es bleiben also 35 M unberührt. Jetzt schauen wir, wie wir unsere Abnahmemaschen gleichmäßig unter diesen 35 M unterbringen. Dafür eignet sich am besten eine "Teilen mit Rest" Rechnung!

```python
quotient, rest = divmod(k, a)
print(f"{k} Maschen geteilt durch {a} Abnahmen ergibt {quotient} Maschen mit {rest} Rest-Maschen")
```

Was bedeutet das? Das heißt, "in der Regel" müssen wir immer 3 M stricken und dann die nächsten 2 M zus. stricken. Aber wir haben noch 2 Rest-Maschen übrig, die wir noch verteilen müssen, und das geht am besten, in dem man 2 mal statt 3 M einfach 4 M am Stück strickt und danach 2 M zus. strickt. Verallgemeinert:

```python
print(f"Wir müssen {rest} mal {quotient + 1} Maschen stricken und die nächsten 2 zus. stricken.")
print(f"Und wir müssen {a - rest} mal {quotient} Maschen stricken und die nächsten 2 zus. stricken.")
```

Im allen Fällen, wo die Zahl der Rest-Maschen gleich 0 ist, sind natürlich die Abstände zwischen den Abnahmen identisch und entsprechen genau dem oben berechneten Quotienten.
